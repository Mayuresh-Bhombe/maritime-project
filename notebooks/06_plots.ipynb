```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\ML project\data\gold\la_port_ml_ready_dataset.csv")

print(df.shape)
df.head()
```

```python
df.info()
df.describe()
```

```python
#PART 1 — OVERALL DATA DISTRIBUTIONS

#Congestion Level Distribution
df["CongestionLevel"].value_counts().plot(kind="bar")
plt.title("Congestion Level Distribution")
plt.xlabel("Congestion Level")
plt.ylabel("Number of Records")
plt.show()
```

```python
#Delay Minutes Distribution
df["DelayMinutes"].plot(kind="hist", bins=50)
plt.title("Delay Minutes Distribution")
plt.xlabel("Delay Minutes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Berth Feasibility Distribution
df["BerthFeasible"].value_counts().plot(kind="bar")
plt.title("Berth Feasibility Distribution")
plt.xlabel("Feasible (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()
```

```python
#Arrival Hour vs Traffic Volume
df["ArrivalHour"].value_counts().sort_index().plot(kind="bar")
plt.title("Traffic Volume by Arrival Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Arrivals")
plt.show()
```

```python
#PART 2 — OPERATIONAL INSIGHTS

#Arrival Hour vs Congestion
df.groupby("ArrivalHour")["CongestionLevel"] \
  .value_counts(normalize=True) \
  .unstack() \
  .plot()

plt.title("Congestion Pattern by Arrival Hour")
plt.ylabel("Proportion")
plt.show()
```

```python
#Terminal-wise Congestion
df.groupby("TerminalID")["CongestionLevel"] \
  .value_counts(normalize=True) \
  .unstack() \
  .plot(kind="bar")

plt.title("Congestion Distribution by Terminal")
plt.ylabel("Proportion")
plt.show()
```

```python
#Vessel Length vs Congestion
df.boxplot(column="VesselLength", by="CongestionLevel")
plt.title("Vessel Length vs Congestion Level")
plt.suptitle("")
plt.xlabel("Congestion Level")
plt.ylabel("Vessel Length")
plt.show()
```
