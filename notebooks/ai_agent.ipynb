```python
import json
import re

def extract_structured_features(user_text):

    prompt = f"""
Return ONLY valid JSON. No explanation. No markdown.

Fields:
ArrivalHour (0-23)
ArrivalDayOfWeek (0=Mon ... 6=Sun)
IsWeekend (0 or 1)
VesselLength
VesselDraft
AvgSOG
BerthDepth
MaxLOA
DailyCapacityTEU
TerminalID (T2,T3,T4,T5,T6,T7)
BerthID (B2,B3,B4,B5,B6,B7,B8,B9,B10)
OperationalStatus (Normal or Maintenance)
BerthTime
WaitingRatio

Scenario:
{user_text}
"""

    text = llm.generate_content(prompt).text

    # extract JSON using regex
    match = re.search(r"\{.*\}", text, re.S)
    if not match:
        raise ValueError("No JSON found in LLM response:\n" + text)

    json_text = match.group()
    return json.loads(json_text)
```

```python
def build_model_inputs(struct):

    rf_df = pd.DataFrame(np.zeros((1, len(rf_features))), columns=rf_features)

    rf_df.loc[0, ["AvgSOG","VesselLength","VesselDraft","ArrivalHour",
                  "ArrivalDayOfWeek","IsWeekend","DailyCapacityTEU",
                  "BerthDepth","MaxLOA"]] = [
        struct["AvgSOG"], struct["VesselLength"], struct["VesselDraft"],
        struct["ArrivalHour"], struct["ArrivalDayOfWeek"], struct["IsWeekend"],
        struct["DailyCapacityTEU"], struct["BerthDepth"], struct["MaxLOA"]
    ]

    rf_df[f"TerminalID_{struct['TerminalID']}"] = 1
    rf_df[f"BerthID_{struct['BerthID']}"] = 1
    rf_df["OperationalStatus"] = 1 if struct["OperationalStatus"]=="Normal" else 0


    lgb_df = pd.DataFrame(np.zeros((1, len(lgb_features))), columns=lgb_features)

    lgb_df.loc[0, ["AvgSOG","VesselLength","VesselDraft","ArrivalHour",
                   "ArrivalDayOfWeek","IsWeekend","BerthDepth",
                   "MaxLOA","DailyCapacityTEU","BerthTime","WaitingRatio"]] = [
        struct["AvgSOG"], struct["VesselLength"], struct["VesselDraft"],
        struct["ArrivalHour"], struct["ArrivalDayOfWeek"], struct["IsWeekend"],
        struct["BerthDepth"], struct["MaxLOA"], struct["DailyCapacityTEU"],
        struct["BerthTime"], struct["WaitingRatio"]
    ]

    lgb_df["DraftMargin"] = struct["BerthDepth"] - struct["VesselDraft"]
    lgb_df["LOAMargin"] = struct["MaxLOA"] - struct["VesselLength"]
    lgb_df["BerthFeasible"] = 1

    lgb_df[f"TerminalID_{struct['TerminalID']}"] = 1
    lgb_df[f"BerthID_{struct['BerthID']}"] = 1
    lgb_df["OperationalStatus_Normal"] = 1 if struct["OperationalStatus"]=="Normal" else 0
    lgb_df["OperationalStatus_Maintenance"] = 1 if struct["OperationalStatus"]=="Maintenance" else 0


    xgb_df = pd.DataFrame(np.zeros((1, len(xgb_features))), columns=xgb_features)

    xgb_df.loc[0, ["VesselLength","VesselDraft","DailyCapacityTEU",
                   "ArrivalHour","ArrivalDayOfWeek","IsWeekend"]] = [
        struct["VesselLength"], struct["VesselDraft"], struct["DailyCapacityTEU"],
        struct["ArrivalHour"], struct["ArrivalDayOfWeek"], struct["IsWeekend"]
    ]

    xgb_df["OperationalStatus"] = 1 if struct["OperationalStatus"]=="Normal" else 0

    return rf_df, lgb_df, xgb_df
```

```python
def predict_from_text(user_text):

    struct = extract_structured_features(user_text)
    rf_df, lgb_df, xgb_df = build_model_inputs(struct)

    congestion = rf_model.predict(rf_df)[0]
    delay = lgb_model.predict(lgb_df)[0]
    berth = xgb_model.predict(xgb_df)[0]

    return {
        "Congestion": congestion,
        "DelayMinutes": round(float(delay),2),
        "BerthFeasible": "Yes" if berth==1 else "No",
        "ExtractedFeatures": struct
    }
```

```python
text = ""
for page in reader.pages:
    text += page.extract_text()
```

```python
def chunk_text(text, chunk_size=800, overlap=100):
    chunks = []
    start = 0
    while start < len(text):
        end = start + chunk_size
        chunks.append(text[start:end])
        start = end - overlap
    return chunks

chunks = chunk_text(text)
print("Chunks:", len(chunks))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(chunks)

print("Vector shape:", X.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_policy_context(query, k=3):

    q_vec = vectorizer.transform([query])
    sims = cosine_similarity(q_vec, X)[0]

    top_idx = np.argsort(sims)[-k:][::-1]
    return "\n".join([chunks[i] for i in top_idx])
```

```python
print(get_policy_context("berth allocation during congestion"))
```

```python
def decision_agent(user_text):

    # ML predictions
    struct = extract_structured_features(user_text)
    rf_df, lgb_df, xgb_df = build_model_inputs(struct)

    congestion = rf_model.predict(rf_df)[0]
    delay = round(float(lgb_model.predict(lgb_df)[0]),2)
    berth = "Yes" if xgb_model.predict(xgb_df)[0]==1 else "No"

    # RAG policy retrieval
    policy = get_policy_context("berth allocation and congestion handling")

    # Final reasoning
    final_prompt = f"""
You are a port decision-support AI working in simulation mode.

User Scenario:
{user_text}

Model Predictions:
Congestion Level: {congestion}
Expected Delay: {delay} minutes
Berth Feasible: {berth}

Relevant Policy:
{policy}

Give final recommendation and explanation clearly.
"""

    return llm.generate_content(final_prompt).text
```

```python
decision_agent(
    "Heavy traffic in evening on Saturday, large vessel, terminal T4 berth B7, normal operation"
)
```

```python
decision_agent(
    "Morning arrival on weekday, small vessel, terminal T2 berth B3, low traffic"
)
```
